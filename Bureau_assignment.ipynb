```python
## Importing the necessary library
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn as sl
from sklearn.feature_selection import SelectKBest, chi2, mutual_info_classif
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
# Read the Datasetof training data
df = pd.read_csv("/content/drive/MyDrive/Assignment_Train.csv")
```

```python
df.head()
```

```python
## Make a list of redundant columns
df_list = ["APPLICATION LOGIN DATE", "HDB BRANCH STATE", "FIRST NAME", "MIDDLE NAME", "LAST NAME", "mobile" , "DEALER NAME", "ASSET CTG", "ASSET MODEL NO", "PRIMARY ASSET MAKE", "Primary Asset Model No", "Personal Email Address", "DOB", "EMPLOYER NAME", "EMPLOYER TYPE", "Pan Name", "name", "upi_name", "Phone Social Premium.a23games", "Phone Social Premium.isWABusiness", "Phone Social Premium.jiomart", "Phone Social Premium.my11", "Phone Social Premium.rummycircle", "Phone Social Premium.whatsapp", "Phone Social Premium.yatra", "phone_digitalage", "phone_nameMatchScore"]
```

```python
# Drop redundant columns
df.drop(labels = df_list, axis=1, inplace = True)
```

```python
df.shape
```

```python
df.head()
```

```python
## Check the missing values in each feature
df.isnull().sum()
```

```python
## Find the mode of Categorical feature HDB Branch name
df["HDB BRANCH NAME"].mode()[0]
```

```python
# Fill all the Blank values with mode
df["HDB BRANCH NAME"].fillna(df["HDB BRANCH NAME"].mode()[0], inplace = True)
```

```python
## Check the number of null values missing value in HDB Branch Name feature
df["HDB BRANCH NAME"].isnull().sum()
```

```python
## For all the object datatype other than Cibil Score and Application status fill all the blank values in respective categorical features  with their mode values.
for i in df.columns:
  if df[i].dtype == "object" and i != "Cibil Score" and i != "Application Status":
    df[i].fillna(df[i].mode()[0], inplace = True)
```

```python
## Check the number of null values/missing values in each feature
df.isnull().sum()
```

```python
## Converting Cibil score to numeric feature
df['Cibil Score']  = pd.to_numeric(df['Cibil Score'], errors = 'coerce')
```

```python
## Filling all the blank values with their mean
df['Cibil Score'].fillna(df['Cibil Score'].mean(), inplace = True)
```

```python
## Check the number of null values in cibil score feature
df['Cibil Score'].isnull().sum()
```

```python
df.dtypes
```

```python
## Converting all the features which start with Phone or phone keyword to object datatype
for i in df.columns:
  if i.startswith("Phone") or i.startswith("phone"):
    df[i] = df[i].astype(object)
```

```python
df.dtypes
```

```python
## Replacing all the blank or null value with the mean of application status
for i in df.columns:
  if df[i].dtype == "object" and i!="Application Status":
    df[i].fillna(df[i].mode()[0], inplace = True)
```

```python
## Replace all the blank or null values of the feature Total Asset Cost with their mean
df["TOTAL ASSET COST"].fillna(df["TOTAL ASSET COST"].mean(), inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

### Label encoding

```python
## Finding the number of unique values of categorical columns
cat_columns = [x for x in df.columns if df[x].dtype=='object']
for i in cat_columns:
  print(f"No of unique values in {i} = {df[i].nunique()}")
```

```python
## Doing label encoding to convert all  the categorical feature into numerical feature
from sklearn.preprocessing import LabelEncoder
lc = LabelEncoder()
one_cat_columns = cat_columns[:-1]
print(one_cat_columns)
for i in one_cat_columns:
  df[i] = lc.fit_transform(df[i])
```

```python
df.dtypes
```

```python
df.dtypes
```

```python
## Converting Mobile Verification feature to integer datatype from object datatype
df['MOBILE VERIFICATION'] = df['MOBILE VERIFICATION'].astype(int)
df.dtypes
```

```python
## Reading all the columns
df.columns
```

```python
## Doing one hot encoding for target feature
df['Application Status'] = pd.get_dummies(df['Application Status'], drop_first=True, dtype=int)
```

```python
df.head()
```

```python
## Spiliting the data into target and non target columns
X = df.drop('Application Status', axis=1).values
y = df['Application Status'].values
```

```python
df.head()
```

```python
## Doing train test split for training data so as to get validation accuracy
from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=1)
```

```python
## Appying standard scaling - This is particularly important when using algorithms that are sensitive to the scale of the data, like SVM, Random Forest
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_valid = sc.transform(X_valid)
```

```python
## Training the data using different models (Logistic Resgression, Random Forest Classifier, Support Vector Machine)
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

rf = RandomForestClassifier(n_estimators=1000)
lr = LogisticRegression(penalty='l2')
svc = SVC(kernel='rbf', degree=7)
lr.fit(X_train, y_train)
rf.fit(X_train, y_train)
svc.fit(X_train, y_train)
```

```python
## Doing evaluation using evaluation metrices
from sklearn.metrics import accuracy_score, precision_score, recall_score

pred = lr.predict(X_train)
pred_lr = rf.predict(X_train)
pred_svc= svc.predict(X_train)
print("Accuracy using logistic regression", accuracy_score(pred, y_train))
print("Precision using logistic regression", precision_score(pred, y_train))
print("Recall using logistic regression",recall_score(pred, y_train))
print("Accuracy using Random Forest", accuracy_score(pred_lr, y_train))
print("Precision using Random Forest",precision_score(pred_svc, y_train))
print("Recall using Random Forest",recall_score(pred_svc, y_train))
print("Accuracy using Support Vector Machine", accuracy_score(pred_svc, y_train))
print("Precision using Support Vector Machine",precision_score(pred_svc, y_train))
print("Recall using Support Vector Machine",recall_score(pred_svc, y_train))


pred = lr.predict(X_valid)
pred_lr = rf.predict(X_valid)
pred_svc= svc.predict(X_valid)
print("Accuracy score using logistic regression", accuracy_score(pred, y_valid))
print("Precision using logistic regression", precision_score(pred, y_valid))
print("Recall using logistic regression",recall_score(pred, y_valid))
print("Accuracy score using Random Forest", accuracy_score(pred_lr, y_valid))
print("Precision using Random Forest", precision_score(pred_lr, y_valid))
print("Recall using Random Forest",recall_score(pred_lr, y_valid))
print("Accuracy score using Support Vector Classifier", accuracy_score(pred_svc, y_valid))
print("Precision using Support Vector Classifier", precision_score(pred_svc, y_valid))
print("Recall using Support Vector Classifier", recall_score(pred_svc, y_valid))
```

```python
## Performing hypperparameter tuning in order to reduce overfitting in case of random forest
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_search = GridSearchCV(RandomForestClassifier(), param_grid, cv=2)
grid_search.fit(X_train, y_train)
print(f"Best parameters: {grid_search.best_params_}")
```

```python
## Doing evaluation using evaluation metrices
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
y_pred = rf.predict(X_train)
y_pred_test = rf.predict(X_valid)
print("Accuracy using Random Forest", accuracy_score(y_pred, y_train))
print("Precision using Random Forest",precision_score(y_pred, y_train))
print("Recall using Random Forest",recall_score(y_pred, y_train))
print("Accuracy score using Random Forest", accuracy_score(y_pred_test, y_valid))
print("Precision using Random Forest", precision_score(y_pred_test, y_valid))
print("Recall using Random Forest",recall_score(y_pred_test, y_valid))
```
